```python
class NodoAVL:
    def __init__(self,elem):
        self.elem=elem
        self.alt=1
        self.izq=None
        self.der=None
        self.papa=None
    
    def getIzq(self):
        return self.izq
    
    def getDer(self):
        return self.der
    
    def getPapa(self):
        return self.papa
    
    def getElem(self):
        return self.elem
    
    def setElem(self,elem):
        self.elem=elem
        
    def getAlt(self):
        return self.alt
```

```python
class ArbolAVL:
    def getAlt(self,nodo):
        if not nodo:
            return 0
        return nodo.alt
    
    def getBal(self,nodo):
        if not nodo:
            return 0
        return self.getAlt(nodo.izq)-self.getAlt(nodo.der)
    
    def rotaDer(self, y):
        x=y.izq
        z=x.der
        x.der=y
        y.izq=z
        y.alt=1+max(self.getAlt(y.izq),self.getAlt(y.der))
        x.alt=1+max(self.getAlt(x.izq),self.getAlt(x.der))
        return x
    
    def rotaIzq(self,x):
        y = x.der
        z = y.izq
        y.izq = x
        x.der = z
        x.alt = 1 + max(self.getAlt(x.izq), self.getAlt(x.der))
        y.alt = 1 + max(self.getAlt(y.izq), self.getAlt(y.der))
        return y
    
    def inserta(self, raiz, elem):
        if not raiz:
            return NodoAVL(elem)
        if elem < raiz.getElem():
            raiz.izq = self.inserta(raiz.izq, elem)
        elif elem > raiz.getElem():
            raiz.der = self.inserta(raiz.der, elem)
        else:
            return raiz
        raiz.alt = 1 + max(self.getAlt(raiz.izq), self.getAlt(raiz.der))
        balance = self.getBal(raiz)
        if balance > 1 and elem < raiz.izq.getElem():
            return self.rotaDer(raiz)
        if balance < -1 and elem > raiz.der.getElem():
            return self.rotaIzq(raiz)
        if balance > 1 and elem > raiz.izq.getElem():
            raiz.izq = self.rotaIzq(raiz.izq)
            return self.rotaDer(raiz)
        if balance < -1 and elem < raiz.der.getElem():
            raiz.der = self.rotaDer(raiz.der)
            return self.rotaIzq(raiz)
        return raiz
    
    def borra(self, raiz, elem):
        if not raiz:
            return raiz

        if elem < raiz.getElem():
            raiz.izq = self.borra(raiz.izq, elem)
        elif elem > raiz.getElem():
            raiz.der = self.borra(raiz.der, elem)
        else:
            if raiz.izq is None:
                temp = raiz.der
                raiz = None
                return temp
            elif raiz.der is None:
                temp = raiz.izq
                raiz = None
                return temp
            temp = self.getMinNodo(raiz.der)
            raiz.setElem(temp.getElem())
            raiz.der = self.borra(raiz.der, temp.getElem())
        raiz.alt = 1 + max(self.getAlt(raiz.izq), self.getAlt(raiz.der))
        balance = self.getBal(raiz)
        if balance > 1 and self.getBal(raiz.izq) >= 0:
            return self.rotaDer(raiz)
        if balance < -1 and self.getBal(raiz.der) <= 0:
            return self.rotaIzq(raiz)
        if balance > 1 and self.getBal(raiz.izq) < 0:
            raiz.izq = self.rotaIzq(raiz.izq)
            return self.rotaDer(raiz)
        if balance < -1 and self.getBal(raiz.der) > 0:
            raiz.der = self.rotaDer(raiz.der)
            return self.rotaIzq(raiz)
        return raiz

    def getMinNodo(self, raiz):
        actual = raiz
        while actual.izq:
            actual = actual.izq
        return actual

    def imprimePorNivel(self, raiz):
        if not raiz:
            return

        nodos = []
        nodos.append(raiz)
        while nodos:
            nivel_actual = len(nodos)
            nodos_nivel = []
            while nivel_actual > 0:
                nodo_actual = nodos.pop(0)
                factor_equilibrio = -self.getBal(nodo_actual)
                nodos_nivel.append(f"{nodo_actual.getElem()}({factor_equilibrio})")
                if nodo_actual.izq:
                    nodos.append(nodo_actual.izq)
                if nodo_actual.der:
                    nodos.append(nodo_actual.der)
                nivel_actual -= 1
            print(",".join(nodos_nivel))
```
